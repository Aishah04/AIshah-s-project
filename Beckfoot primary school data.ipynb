```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('BeckfootSensor-A.csv')
data.head()
```

```python
data.plot(x='created_at', y=['PM1.0_CF1A_ug/m3', 'PM1.0_CF1B_ug/m3'])
plt.xlabel('Date (2021)')
plt.ylabel('Particulate matter (ug/m3)')
plt.title('PM1 levels at Beckfoot Primary School')
plt.tight_layout()

plt.savefig('PM1Beckfootplot.jpeg')
plt.show()
```
